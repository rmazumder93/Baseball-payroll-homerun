```python
from kaggle.api.kaggle_api_extended import KaggleApi
api = KaggleApi()
api.authenticate()
api.dataset_download_files('omipelcastre/mlb-team-statistics')
import zipfile
with zipfile.ZipFile('mlb-team-statistics.zip', 'r') as zip_ref:
    zip_ref.extractall()
api.dataset_download_files('mathurinache/top10totalmlbteampayrolls')
import zipfile
with zipfile.ZipFile('top10totalmlbteampayrolls.zip', 'r') as zip_ref:
    zip_ref.extractall('')
import pandas as pd
```

```python
mlb_data = pd.read_csv('mlb_teams.csv')
mlb_data
```

```python
payroll_data = pd.read_csv('top 10 total MLB team payrolls.csv')
payroll_data
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
mlb_stat = pd.DataFrame(mlb_data, columns=['TeamName','HR'])
payroll = pd.DataFrame(payroll_data, columns=['Team Name', '2017', '2018'])
#Cleaning the Data
mlb_17 = mlb_stat[mlb_stat['TeamName'].str.contains('2017')]
mlb_18 = mlb_stat[mlb_stat['TeamName'].str.contains('2018')]
missing_data = pd.DataFrame(payroll.isna().sum())
missing_payroll = payroll.dropna()['Team Name'].count()
team_payroll = payroll.dropna()
print('Total amount of team with sufficient data is', missing_payroll)
print('Total team missing and removed payroll info for year 2017 is',missing_data.iloc[1,0],'and 2018 is', missing_data.iloc[2,0])
```

```python
#renaming the columns and filtering only the team that have team payroll data aviable for 2017 & 2018
team_payroll = team_payroll.rename(columns={"2017":"PAYROLL'17","2018":"PAYROLL'18"}) 
yr17 ='2017 '
team = [yr17 +'NYY', yr17+'LAD', yr17+'BOS',yr17+'TOR',yr17+'SEA',yr17+'SFG',yr17+'CHC',yr17+'WSN']
mlb_stat_new17 = mlb_17[mlb_17['TeamName'].isin(team)]
yr18 = '2018 '
team = [yr18 +'NYY', yr18+'LAD', yr18+'BOS',yr18+'TOR',yr18+'SEA',yr18+'SFG',yr18+'CHC',yr18+'WSN']
mlb_stat_new18 = mlb_18[mlb_18['TeamName'].isin(team)]
mlb_stat_new18['TeamName'][3] = team_payroll['Team Name'][7]
mlb_stat_new18['TeamName'][4] = team_payroll['Team Name'][16]
mlb_stat_new18['TeamName'][13] = team_payroll['Team Name'][1]
mlb_stat_new18['TeamName'][18] = team_payroll['Team Name'][0]
mlb_stat_new18['TeamName'][23] = team_payroll['Team Name'][12]
mlb_stat_new18['TeamName'][24] = team_payroll['Team Name'][13]
mlb_stat_new18['TeamName'][28] = team_payroll['Team Name'][11]
mlb_stat_new18['TeamName'][29] = team_payroll['Team Name'][23]
mlb_stat_new17['TeamName'][33] = team_payroll['Team Name'][7]
mlb_stat_new17['TeamName'][34] = team_payroll['Team Name'][16]
mlb_stat_new17['TeamName'][43] = team_payroll['Team Name'][1]
mlb_stat_new17['TeamName'][48] = team_payroll['Team Name'][0]
mlb_stat_new17['TeamName'][53] = team_payroll['Team Name'][12]
mlb_stat_new17['TeamName'][54] = team_payroll['Team Name'][13]
mlb_stat_new17['TeamName'][58] = team_payroll['Team Name'][11]
mlb_stat_new17['TeamName'][59] = team_payroll['Team Name'][23]
```

```python
#replace the coloumn name to its specific year stat 
mlb17=mlb_stat_new17.rename(columns={"HR":"HR'17"})
mlb18=mlb_stat_new18.rename(columns={"HR":"HR'18"})
team_payroll = team_payroll.rename(columns={'Team Name':'TeamName'})
team_payroll = pd.merge(mlb17,team_payroll, on='TeamName')
mlb_data = pd.merge(mlb18, team_payroll, on='TeamName')
mlb_data = mlb_data.set_index('TeamName')
```

```python
#Converting str to int to run statsitical analysis 
mlb_data["PAYROLL'17"] = mlb_data["PAYROLL'17"].str.replace(",","").astype(int)
mlb_data["PAYROLL'18"] = mlb_data["PAYROLL'18"].str.replace(",","").astype(int)
```

```python
#clean data
mlb_data
```

```python
fig1 = plt.figure()
ax1=fig1.add_subplot(2,1,1)
ax2=fig1.add_subplot(2,1,2)
ax1.set_xlabel("HR'17")
ax1.set_ylabel("PAYROLL'17")
ax1.set_title("HR'17 vs PAYROLL'17")
ax2.set_title("HR'18 vs PAYROLL'18")
ax2.set_xlabel("HR'18")
ax2.set_ylabel("PAYROLL'18")
ax1.plot(mlb_data["HR'17"],mlb_data["PAYROLL'17"], 'o')
ax2.plot(mlb_data["HR'18"],mlb_data["PAYROLL'18"], 'o')
plt.tight_layout()
_18hr_payroll = mlb_data.corr()["PAYROLL'18"]["HR'18"]
_17hr_payroll = mlb_data.corr()["PAYROLL'17"]["HR'17"]
print("According to the analysis there is a negative corrleation between payroll and HR, meaning higher payoll can lead to a lower Homerun as seen by 2017 correlation",format(_17hr_payroll,'.2f'),"and 2018 correlation",format(_18hr_payroll,".2f"))
```

```python
_17max_payroll = mlb_data["PAYROLL'17"].max()
_17min_payroll = mlb_data["PAYROLL'17"].min()
_18max_payroll = mlb_data["PAYROLL'18"].max()
_18min_payroll = mlb_data["PAYROLL'18"].min()
print('Team with the highest payroll in 2017 is' , "${:,}".format(_17max_payroll),'and the lowest payroll is', "${:,}".format(_17min_payroll)+'.')
print('Team with the highest payroll in 2018 is', "${:,}".format(_18max_payroll),'and the lowest payroll is',"${:,}".format(_18min_payroll)+'.')
describe = mlb_data[["PAYROLL'17","PAYROLL'18"]].describe()
payroll17_std = "${:,.2f}".format(describe["PAYROLL'17"]['std'])
payroll18_std = "${:,.2f}".format(describe["PAYROLL'18"]['std'])
print('Standard deviation for payroll for 2017 is',payroll17_std)
print('Standard deviation for payroll for 2018 is',payroll18_std)
ax3 = mlb_data[["HR'18","HR'17"]].plot(kind='bar', stacked ='True',xlabel='Team', ylabel='Homerun', title = 'Total Homeruns for 17 & 18')
ax4 = mlb_data[["PAYROLL'18","PAYROLL'17"]].plot(kind='bar', stacked='True',xlabel='Team', ylabel='Payroll in Hundred Millions', title="Total Payroll for 17 & 18")
plt.tight_layout()
```

```python
#Exporting all the clean data including Statistical analysis including correlation between homerun and payroll 
mlb_pr17_std = mlb_data["PAYROLL'17"].std()
mlb_pr18_std = mlb_data["PAYROLL'18"].std()
df ={"Payroll'17":[_17max_payroll,_17min_payroll,mlb_pr17_std,_17hr_payroll],"Payroll'18":[_18max_payroll,_18min_payroll,mlb_pr18_std,_18hr_payroll], " ":['Max','Min','STD', 'Corr/HR']}
df = pd.DataFrame(df)
df.index=df[' ']
df =df.drop([' '], axis = 1)
export = pd.concat([df,mlb_data])
export.to_csv('Rezwan_Mazumder_CIS9650_Final_Project.csv', header=True)
```
